```python
#This takes 2 - 3 days to run due to rtree.
bins = np.zeros(33)
lists = np.zeros(33)
rtrees = np.zeros(33)
i = 0
for d in directories:
    for sp in shpname:
        testfile = d + sp
        shp = ps.open(testfile)
        t1 = time.time()
        wbin = binW(shp, QUEEN)
        t2 = time.time()
        bins[i] = t2 - t1
        t1 = time.time()
        wlist = listW(shp, QUEEN)
        t2 = time.time()
        lists[i] = t2 - t1
        t1 = time.time()
        wrtree = rtreeW(shp, QUEEN)
        t2 = time.time()
        rtrees[i] = t2 - t1
        try:
            assert wbin.w == wlist.w == wrtree.w
        except:
            print "FAILED on {}".format(testfile)
        print "Bin:{}  List:{}  RTree:{}".format(bins[i], lists[i], rtrees[i])  
        i += 1
```
