```python
samplesizes = [32, 64, 128, 160, 192, 256, 288, 320, 384, 448, 512]
```

```python
#Triangle Lattice
def build_lattice_shapefile(nrows, ncols, outFileName):
    """  
    Build a lattice shapefile with nrows rows and ncols cols

    Parameters
    ----------

    nrows       : int
                  Number of rows
    ncols       : int
                  Number of cols
    outFileName : str
                  shapefile name with shp suffix

    Returns
    -------
    None
    """
    if not outFileName.endswith('.shp'):
        raise ValueError("outFileName must end with .shp")
    o = ps.open(outFileName, 'w') 
    dbf_name = outFileName.split(".")[0] + ".dbf"
    d = ps.open(dbf_name, 'w') 
    d.header = [ 'ID' ]
    d.field_spec = [ ('N', 8, 0) ] 
    c = 0
    for i in xrange(nrows/2):
        for j in xrange(ncols/2):
            ll = i, j 
            ul = i, j + 1
            ur = i + 1, j + 1
            lr = i + 1, j 
            cen = i + 0.5, j + 0.5
            
            t1 = [lr, cen, ll, lr]
            t2 = [ll, cen, ul, ll]
            t3 = [ul, cen, ur, ul]
            t4 = [ur, cen, lr, ur]
            tris = [t1, t2, t3, t4]
            for t in tris:
                o.write(ps.cg.Polygon([t]))
                d.write([c])
                c += 1 
    d.close()
    o.close()
for s in samplesizes:
    build_lattice_shapefile(s, s, 'trianglelattices/{}.shp'.format(s*s))
```
